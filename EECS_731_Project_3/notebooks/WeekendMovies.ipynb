```python
import pandas as pd
import numpy as np
import sklearn as sc
from matplotlib import pyplot as plt

#loading all four datasets into dataframes
df_links = pd.read_csv('../data/raw/links.csv')
df_movies = pd.read_csv('../data/raw/movies.csv')
df_ratings = pd.read_csv('../data/raw/ratings.csv')
df_tags = pd.read_csv('../data/raw/tags.csv')
```

```python
df_movies.head()
```

Now I'm going to merge the 4 dataframes. Earlier I tried to remerge on the same data frame but Python didn't like that for some reason so I'll merge on a new dataframe for each merge (even though I'm sure there is a better way to accomplish this). 

```python
df_merged = pd.merge(df_movies, df_links, on='movieId')
df_merged.head()
```

```python
df_merged1 = pd.merge(df_merged, df_ratings, on = 'movieId')
df_merged1.head()
```

```python
df_merged2 = pd.merge(df_merged1, df_tags, on = 'movieId')
#df_merged2 has all features
df_merged2.head()
```
